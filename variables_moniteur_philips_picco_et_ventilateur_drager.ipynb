```python
import pandas as pd

# Load the Excel file to see the content
file_path = r"D:\stage\data\Detail_Courbes et Dispositifs_CH Saint Denis_20211001.xlsx"
excel_data = pd.ExcelFile(file_path)

# Get the names of all the sheets in the Excel file
sheet_names = excel_data.sheet_names
sheet_names
```

```python
# Récupérer les 3 feuilles  de calculs :  'Moniteurs Philips', 'Moniteur PICCO 2','Respirateurs V500&V800'
# On récupère chaque feuille dans un dataframe distinct

data_philips = pd.read_excel(r"D:\stage\data\Detail_Courbes et Dispositifs_CH Saint Denis_20211001.xlsx", sheet_name='Moniteurs Philips')

data_pulsion = pd.read_excel(r"D:\stage\data\Detail_Courbes et Dispositifs_CH Saint Denis_20211001.xlsx", sheet_name='Moniteur PICCO 2')

data_drager = pd.read_excel(r"D:\stage\data\Detail_Courbes et Dispositifs_CH Saint Denis_20211001.xlsx", sheet_name='Respirateurs V500&V800')
```

```python
# Données Moniteur philips
data_philips.head()
```

```python
data_philips.columns
```

```python
# Extraire les colonnes nécessaires
colonnes = ['ID', 'Label']
data = data_philips[colonnes].assign(Unit=pd.NA, Device='Philips').sort_values(by='Label') # ajouter des colonnes unit et Device, puis trier les valeurs au même temps

display(data)
```

```python
# Données pulsion (Picco 2)
data_pulsion.head()
```

```python
# Extraire les trois colonnes id, label et unit, puis concaténer le résultat avec data
data = pd.concat([data, data_pulsion[['ID', 'Label', 'Unit']].assign(Device='Pulsion')], ignore_index=True)
display(data)
```

```python
# Données Drager ('Respirateurs V500&V800')
display(data_drager)
```

```python
data = pd.concat([data, data_drager[['ID', 'Label', 'Unit']].assign(Device='drager')])
display(data)
```

```python
# Ajouter une colonne format avec comme valeur Waveforms
data = data.assign(Format='WaveForms')
display(data)
```

```python
# sauvegarder sous format excel et csv
data.to_excel(r"C:\Users\AdminPC\sepsis_prediction\sepsis_prediction\exploration du sujet\variables_wave.xlsx", index=False)
data.to_csv(r"C:\Users\AdminPC\sepsis_prediction\sepsis_prediction\exploration du sujet\variables_wave.csv", index=False)
```
