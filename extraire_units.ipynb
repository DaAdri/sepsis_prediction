```python
import pandas as pd

# Define the path to your text file
file_path = r"C:\Users\AdminPC\sepsis_prediction\sepsis_prediction\data\Resultat test 2406\Resultat test 2406\MX800\MX800.txt"

# Initialize an empty list to store the extracted data
data = []

# Open and read the text file
with open(file_path, 'rb') as file:
    content = file.read()

# Decode content with utf-8 and ignore errors
decoded_content = content.decode('utf-8', errors='ignore')

# Split the decoded content into lines
lines = decoded_content.split('\n')

for line in lines:
    # Split the line by ';'
    parts = line.strip().split(';')
    if len(parts) > 1:
        ID = parts[0]
        Label_ok = parts[1]
        id_unit = parts[-3] if len(parts) > 3 else ''
        unit = parts[-2] if len(parts) > 2 else ''
        # Append the extracted information to the data list
        data.append([ID, Label_ok, id_unit, unit])

# Create a DataFrame from the extracted data
df = pd.DataFrame(data, columns=['ID', 'Label_ok', 'id_unit', 'unit'])

# Display the DataFrame
display(df.shape)

```

```python
# Fonction pour supprimer l'astérisque à la fin et convertir en entier
def clean_id(value):
    if isinstance(value, str) and value.endswith('*'):
        return int(value[:-1])
    return int(value)

# Appliquer la fonction à la colonne 'ID'
df_units['ID'] = df_units['ID'].apply(clean_id)
```

```python
df_units['id_unit'] = df_units['id_unit'].astype(int)
df_units = df_units.drop_duplicates(subset='ID')
display(df_units)
```
