```python
# Une fois ce notebook exécuté la première fois, aller directement à la cellure qui charge les dfs déja sauvegardés


import pandas as pd

# Load the Excel file to see the content
file_path = r"C:\Users\AdminPC\sepsis_prediction\sepsis_prediction\data\Variables des DM_Perimètre Actuel_CH Delafontaine_20240111.xlsx"
excel_data = pd.ExcelFile(file_path)

# Get the names of all the sheets in the Excel file
sheet_names = excel_data.sheet_names
sheet_names
```

```python
# Charger chaque feuille de calcule dans un dataframe, avec le nom de la feuille suivit de _data comme nom du df
# Créer un dictionnaire pour sauvegarde les dataframes
dataframes = {}

# Load each sheet into a dataframe and store it in the dictionary
for sheet in sheet_names:
    df_name = f"{sheet}_data"
    dataframes[df_name] = pd.read_excel(file_path, sheet_name=sheet)

# Display the names of the dataframes created
print(list(dataframes.keys()))
```

```python
display(dataframes['Artis_data'].columns)
```

```python
# Charger les feuilles de calcules contenant des variables dans des df distincts

# Columns to check for
required_columns = ['ID', 'Label', 'Unit', 'Group']

# Create a dictionary to hold the filtered dataframes
filtered_dataframes = {}

# Iterate over the dataframes
for name, df in dataframes.items():
    # Strip whitespace from the column names
    df.columns = df.columns.str.strip()
    
    # Check if the dataframe contains the required columns
    if set(required_columns).issubset(df.columns):
        # Select only the required columns
        filtered_df = df[required_columns]
        
        # Supprimer les lignes vides
        filtered_df.dropna(inplace=True)
        
        # Remettre le type de la colonne ID en int
        filtered_df['ID'] = filtered_df['ID'].astype(int)
        
        # Remove spaces from the dataframe name
        new_name = name.replace(' ', '_')
        filtered_dataframes[new_name] = filtered_df

# Display the names of the filtered dataframes
print(filtered_dataframes.keys())
```

```python
# Affichage des dataframes résultants
for name, df in filtered_dataframes.items():
    display(name, df)
```

```python
# Sauvegarde des dataframes en format csv et excel
for name, df in filtered_dataframes.items():
    df.to_csv(f"C:/Users/AdminPC/sepsis_prediction/sepsis_prediction/exploration du sujet/variables/{name}.csv", index=False)
```

```python
import os
import pandas as pd

# Dossier contenant les fichiers CSV
folder_path = f"C:/Users/AdminPC/sepsis_prediction/sepsis_prediction/exploration du sujet/variables"

# Initialiser un dictionnaire pour stocker les DataFrames
dataframes = {}

# Initialiser un ensemble pour stocker tous les IDs
all_ids = set()

# Obtenir la liste des fichiers CSV dans le dossier
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Charger chaque fichier CSV dans une DataFrame et l'assigner à une variable
for file_name in csv_files:
    # Créer le nom de la variable en utilisant le nom du fichier sans extension
    key_name = os.path.splitext(file_name)[0]
    
    # Remplacer les caractères non autorisés dans le nom de la variable
    key_name = key_name.replace(' ', '_').replace('-', '_')
    
    # Construire le chemin complet du fichier CSV
    file_path = os.path.join(folder_path, file_name)
    
    # Charger le fichier CSV dans une DataFrame
    df = pd.read_csv(file_path)
    
     # Ajouter la DataFrame au dictionnaire
    dataframes[key_name] = df
    
     # Ajouter les IDs à l'ensemble
    all_ids.update(df['ID'])
    
    
```

```python
# voir s'il y a des variables qui sont présentes dans toutes les DFs  (Intersection)

# Commencer avec la première DataFrame pour les jointures successives
result_df = None

for key, df in dataframes.items():
    if result_df is None:
        result_df = df
    else:
        result_df = result_df.merge(df[['ID']], on='ID', how='inner')


print(result_df)
```
