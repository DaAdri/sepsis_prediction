```python
var_data = df
global_data = pd.read_excel(r"C:\Users\AdminPC\sepsis_prediction\sepsis_prediction\exploration du sujet\variables\variables_distinctes_global.xlsx")
# Function to find partial matches
def find_partial_matches(label, df2):
    return df2[df2['Label'].str.contains(label, case=False, na=False)]['ID'].tolist()

# Fonction pour traiter les labels composés
def process_label(label, df2):
    parts = label.split()
    matches = []
    matches.extend(find_partial_matches(parts[0], df2))
    #for part in parts:
    #    matches.extend(find_partial_matches(part, df2))
    return list(set(matches))

# Appliquer la fonction sur chaque colonne indépendamment
var_data['ids_label'] = var_data['Label'].apply(lambda x: process_label(x, global_data))
# appliquer la fonction à l'intégralité du Label
# var_data['ids_label'] = var_data['Label'].apply(lambda x: find_partial_matches(x, global_data))

var_data['ids_variable'] = var_data['variable'].apply(lambda x: find_partial_matches(x, global_data))

# Combiner les résultats et éliminer les doublons
var_data['ids_possible'] = var_data.apply(lambda row: list(set(row['ids_label'] + row['ids_variable'])), axis=1)

# Créer un nouveau DataFrame avec les résultats combinés
result_data = {
    'Label': var_data['Label'],
    'variable': var_data['variable'],
    'Unit' : var_data['Unit'],
    'ids_possible': var_data['ids_possible']
}

result_df = pd.DataFrame(result_data)
result_df
```

```python
# Ajouter la colonne 'autres_appelations'
def find_autres_appelations(ids, df):
    return df[df['ID'].isin(ids)]['Label'].tolist()

result_df['appelations_possible'] = result_df['ids_possible'].apply(lambda ids: find_autres_appelations(ids, global_data))

result_df
```
