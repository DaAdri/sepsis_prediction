```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_file = 'Dataset.csv'
data = pd.read_csv(data_file)
data.drop(columns=['Unnamed: 0', 'Unit1', 'Unit2'], inplace=True)
data.head()
```

```python
# Histogrammes pour les variables numériques
num_features = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
data[num_features].hist(bins=15, figsize=(20, 20), layout=(9, 5))
plt.show()
```

```python
# Diagrammes en boîte pour visualiser les distributions et identifier les outliers
for column in num_features:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

```

```python
# Matrice de corrélation
plt.figure(figsize=(20, 16))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()

```

```python
# Diagrammes en barres pour les variables catégoriques
cat_features = data.select_dtypes(include=['object']).columns.tolist() + ['Gender', 'SepsisLabel']
for column in cat_features:
    plt.figure(figsize=(10, 4))
    sns.countplot(x=column, data=data)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Tri des données par patient et heure
data.sort_values(by=['Patient_ID', 'Hour'], inplace=True)
```

```python
# Nombre de patient atteint du sepsis
data['Patient_ID'][data['SepsisLabel'] == 1].value_counts()
```

```python
# Sous df composé des patients atteints de sepsis au cours de leur séjour
sepsis_patients = data['Patient_ID'][data['SepsisLabel'] == 1]
df_sepsis_patients = data[data['Patient_ID'].isin(sepsis_patients)]
df_sepsis_patients
```

```python
# # Sous df composé des patients non atteints de sepsis au cours de leur séjour
df_non_sepsis_patients = data[~data['Patient_ID'].isin(sepsis_patients)]
df_non_sepsis_patients
```

```python
# Labellisation pour la classification et la régression
df = data.copy()
df['will_have_sepsis'] = df.groupby('Patient_ID')['SepsisLabel'].transform(lambda x: 1 if x.max() > 0 else 0)
df['time_to_sepsis'] = df.apply(lambda row: (df[(df['Patient_ID'] == row['Patient_ID']) & (df['SepsisLabel'] == 1)]['Hour'].min() - row['Hour']) if row['will_have_sepsis'] else np.nan, axis=1)

# Filtrer les données pour entraînement (suppression des nan si nécessaire)
train_data = df.dropna(subset=['time_to_sepsis'])
train_data
```

```python
# train_data.to_excel('data_cleaned.xlsx', index=False)

train_data = pd.read_excel('data_cleaned.xlsx')
```

```python
df = train_data.copy()
# Aplatir le datset
# df = data.drop(columns='ICULOS')
# display(df)
temporelle_columns = ['HR', 'O2Sat', 'Temp', 'SBP', 'MAP', 'DBP', 'Resp', 'EtCO2',
       'BaseExcess', 'HCO3', 'FiO2', 'pH', 'PaCO2', 'SaO2', 'AST', 'BUN',
       'Alkalinephos', 'Calcium', 'Chloride', 'Creatinine', 'Bilirubin_direct',
       'Glucose', 'Lactate', 'Magnesium', 'Phosphate', 'Potassium',
       'Bilirubin_total', 'TroponinI', 'Hct', 'Hgb', 'PTT', 'WBC',
       'Fibrinogen', 'Platelets', 'SepsisLabel', 'will_have_sepsis', 'time_to_sepsis']

# Fonction pour transformer les colonnes en listes
def transform_to_list(group):
    transformed = {}
    for col in temporelle_columns:
        transformed[col] = group[col].tolist()  # Convertir la colonne en liste
    transformed['Hour'] = group['Hour'].tolist()  # Assurez-vous d'inclure l'heure pour le suivi
    return pd.Series(transformed)

# Appliquer la transformation
transformed_df = df.groupby('Patient_ID').apply(transform_to_list)
```
