```python
import hl7apy
from hl7apy import parser
from hl7apy.core import Group, Segment
from hl7apy.exceptions import UnsupportedVersion

# Chemin vers le fichier HL7
file_path1_1 = r"D:\stage\data\archive(6)\EXEMPLE1_GLIMS_HL7_RESULTATS\EXEMPLE1_GLIMS_HL7_RESULTATS\2404161450-0001.txt"
file_path1_2 = r"D:\stage\data\archive(6)\EXEMPLE1_GLIMS_HL7_RESULTATS\EXEMPLE1_GLIMS_HL7_RESULTATS\2404161450-0002.txt"
file_path2_1 = r"D:\stage\data\archive(6)\EXEMPLE1_GLIMS_HL7_RESULTATS\EXEMPLE1_GLIMS_HL7_RESULTATS\2404161510-0001.txt"
file_path2_2 = r"D:\stage\data\archive(6)\EXEMPLE1_GLIMS_HL7_RESULTATS\EXEMPLE1_GLIMS_HL7_RESULTATS\2404161515-0001.txt"

# Lecture du contenu du fichier
with open(file_path1_1, "r") as file:
    hl7 = file.read()

# Tentative d'analyse du message HL7
try:
    msg1_1 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
except UnsupportedVersion:
    print('Unsupported version')
    msg1_1 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
###############    
# Lecture du contenu du fichier
with open(file_path1_2, "r") as file:
    hl7 = file.read()

# Tentative d'analyse du message HL7
try:
    msg1_2 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
except UnsupportedVersion:
    print('Unsupported version')
    msg1_2 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)

###############    
# Lecture du contenu du fichier
with open(file_path2_1, "r") as file:
    hl7 = file.read()

# Tentative d'analyse du message HL7
try:
    msg2_1 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
except UnsupportedVersion:
    print('Unsupported version')
    msg2_1 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
    
###############    
# Lecture du contenu du fichier
with open(file_path2_2, "r") as file:
    hl7 = file.read()

# Tentative d'analyse du message HL7
try:
    msg2_2 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
except UnsupportedVersion:
    print('Unsupported version')
    msg2_2 = parser.parse_message(hl7.replace('\n', '\r'), find_groups=True, validation_level=2)
    
    
    
# Indentations pour l'affichage
indent = "    "
indent_seg = "    "
indent_fld = "        "

def subgroup(group, indent):
    indent = indent + "    "
    print(indent, group)
    for group_segment in group.children:
        if isinstance(group_segment, Group):
            subgroup(group_segment, indent)
        else:
            print(indent_seg, indent, group_segment)
            for attribute in group_segment.children:
                print(indent_fld, indent, attribute, attribute.value)

def showmsg(msg):
    for segment in msg.children:
        if isinstance(segment, Segment):
            print(indent, segment)
            for attribute in segment.children:
                print(indent_fld, indent, attribute, attribute.value)
        elif isinstance(segment, Group):
            print(indent, segment)
            for group_segment in segment.children:
                if isinstance(group_segment, Group):
                    subgroup(group_segment, indent)
                else:
                    print(indent_seg, indent, group_segment)
                    for attribute in group_segment.children:
                        print(indent_fld, indent, attribute, attribute.value)
```
